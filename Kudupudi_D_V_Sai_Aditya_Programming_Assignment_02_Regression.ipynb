```python
#Understanding Error Surface
#All imports
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import collections

########################################
#Generate meaningfull data
########################################
np.random.seed(1000)
heights=np.random.uniform(50,80,25)
weights=3.86*heights-110.42+np.random.normal(loc=0,scale=20**0.5,size=25)
# print(heights)
# print(weights)



########################################
#Plot scatter plot of data
########################################
plt.scatter(heights,weights)
plt.xlabel('Heights')
plt.ylabel('Weights')
plt.title("Data")

########################################
#Weight estimation through error surface, i.e., empirically locate the minima of error surface
########################################
#Sample a bunch of w's around w_opt and compute the associated error

#Compute the error
def Error(w,t,x): #inputs : 1)weight 2)data i.e (t,x)
    #Estimate the target
    y=w[0]+w[1]*x
    #Compute and return the error
    error=np.sum((y-t)**2)
    return error

samples=1000
opt=[-110.42,3.86]

w0=np.linspace(opt[0]-50,opt[0]+50,samples)
w1=np.linspace(opt[1]-1.5,opt[1]+1.5,samples)

W0,W1=np.meshgrid(w0,w1)
errors = np.empty_like(W0)

for i in range(samples):
    for j in range(samples):
        w=[w0[i],w1[j]]
        errors[j, i] = Error(w, weights, heights)

#Plot 3D error surface and the corresponding contour plots
#Error surface is a function of w0 and w1
min_error_idx = np.unravel_index(np.argmin(errors), errors.shape)
min_error = errors[min_error_idx]
optimal_w0 = w0[min_error_idx[1]]
optimal_w1 = w1[min_error_idx[0]]
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W0, W1, errors, cmap='viridis')
ax.set_xlabel('w0')
ax.set_ylabel('w1')
ax.set_zlabel('Error')
ax.set_title('Error Surface J(w0, w1)')
ax.scatter(optimal_w0, optimal_w1, min_error, color='black', label='Minimum Error')
# ax.annotate('optimal point',xy=(optimal_w0,optimal_w1))
ax.legend()
plt.show()



# Plot the minimum error point



#Locate the minima of the error surface
print("Minimum Error:", min_error)
print("Optimal w0:", optimal_w0)
print("Optimal w1:", optimal_w1)

# Plot contour graph of the error surface

plt.figure(figsize=(10, 6))
plt.contour(W0, W1, errors, levels=20, cmap='viridis')
plt.xlabel('w0')
plt.ylabel('w1')
plt.title('Contour Plot of Error Surface')
plt.colorbar(label='Error')
plt.scatter(optimal_w0, optimal_w1, color='red', label='Minimum Error', zorder=5)
plt.annotate('Minimum Error', xy=(optimal_w0, optimal_w1), xytext=(optimal_w0 + 10, optimal_w1 + 0.5),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
########################################
#Least squares approach to estimate the weights
########################################
#Complete the below linear regression function
def LinearRegression(x,t): #inputs : 1)input data i.e (x). 2)target i.e (t)
    X=np.array(list(zip([1]*25,x)))
    xtx=np.dot(X.T,X)
    xtx_inv = np.linalg.inv(xtx)
    w_opt = np.dot(np.dot(xtx_inv, X.T), t)
    return w_opt


#Estimate optimal weight's using "LinearRegression" function
w_opt=LinearRegression(heights,weights)
print(w_opt)


#Estimate the targets using the input x and the estimated weights
weights_est=w_opt[0]+w_opt[1]*heights


#Plot the estimated line on top of the above scatter plot
fig,ax=plt.subplots()
ax.scatter(heights,weights,color='black')
ax.plot(heights,weights_est,label='Linear Regression',color='red')
ax.legend()
plt.show()

########################################
#Compare the estimated weight's using least squares approach with the error surface approach
########################################



```

```python
#Understanding model order and overfitting
########################################
#Generate 20 data points
########################################
np.random.seed(1000)
p=2*np.pi
x=np.random.rand(20)
y=np.sin(p*x)+np.random.normal(loc=0,scale=0.1**0.5,size=20)


#Obtain train and test splits

x_train=x[:10]
y_train=y[:10]
x_test=x[10:]
y_test=y[10:]

########################################
#Fitting Mth degree polynomial using least squares approach
########################################
#Complete the function
def PolynomialFit(X_train,Y_train,M,lamda): #(training data, training targets, Model order, Regularization coefficient)
    #Transform the data using polynomial kernel
    X=[]
    for i in range(len(X_train)):
        x=[]
        for j in range(M+1):
            x.append(X_train[i]**j)
        X.append(x)
    X=np.array(X)
#     print(np.dot(X.T,X)+lamda*np.identity(M+1))
    #Find Pseudo inverse solution

    xtx=np.dot(X.T,X)+lamda*np.identity(M+1)
    xtx_inv = np.linalg.inv(xtx)
    w_opt = np.dot(np.dot(xtx_inv, X.T), Y_train)
#     print(w_opt)
#     return the weight vector
    return w_opt


#Complete the function
def PolynomialPred(w_est,X_train,X_test): #(weight,training data, testing data, training targets, testing targets)
    #Estimate the targets for both training and testing data
    phi_train=np.vander(X_train,len(w_est),increasing=True)
    phi_test=np.vander(X_test,len(w_est),increasing=True)
    Train_pred=np.dot(phi_train,w_est)
    Test_pred=np.dot(phi_test,w_est)
#     print(Y_train)
#     print(Y_test)


    #Return training and testing predictions
    return Train_pred,Test_pred

#Complete the function
def PolynomialPred_Error(w_est,X_train,X_test,Y_train,Y_test): #(weight,training data, testing data, training targets, testing targets)
    #Estimate the targets for both training and testing data
    Train_pred,Test_pred=PolynomialPred(w_est,X_train,X_test)
    TrainError=np.sum((Train_pred-Y_train)**2)
    TestError=np.sum((Test_pred-Y_test)**2)

    #Return training and testing error
    return TrainError,TestError

#Iterate through range of M values
M_range=list(range(10))
x_r=np.linspace(0,1,100)
y_r=np.sin(p*x_r)
TrError = []
TeError = []
Train_Est=[]
Test_Est=[]
y_real_p=[]
for M in M_range:

    #Fit Mth order polynomial i.e estimate optimal w. Use the function "PolynomialFit"
    w_est=PolynomialFit(x_train,y_train,M,0)

    #Predict training and testing targets
    Train_est,Test_est=PolynomialPred(w_est,x_train,x_test)
    Train_Est.append(Train_est)
    Test_Est.append(Test_est)
    real,_=PolynomialPred(w_est,x_r,x_r)
    y_real_p.append(real)
    #Predict errors on both training and testing data using estimated w. Use the function "PolynomialPred_Error"
    Train_error,Test_error=PolynomialPred_Error(w_est,x_train,x_test,y_train,y_test)
    TrError.append(Train_error)
    TeError.append(Test_error)
    #Store them for plotting
# print(y_real_p)

#Plot training and testing estimates alogwith the original targets
# print(TeError,TrError)
# fig,ax=plt.subplots(nrows=5,ncols=2)

for i in range(3):
    fig,ax=plt.subplots(1,3,figsize=(12,4),dpi=300)
    fig.suptitle("$sin(2πx)$,no of points=20")
    for j in range(3):
        M=3*i+j
        ax[j].scatter(x_train,y_train,label='Training Data',color='green')
        ax[j].scatter(x_train,Train_Est[3*i+j],label='Predicted Training Data',color='lightgreen',alpha=0.5)
        ax[j].scatter(x_test,y_test,label='Testing Data',color='steelblue')
        ax[j].scatter(x_test,Test_Est[3*i+j],label="Predicted Testing Data",color='skyblue',alpha=0.5)
        ax[j].set_title(f"Plot for M={M+1}")
        ax[j].plot(x_r,y_r,color='crimson')
        ax[j].plot(x_r,y_real_p[3*i+j],color='teal')
        ax[j].set_ylim(-3,3)
        ax[j].legend()
    plt.show()

#Plot training error vs polynomial order, and testing error vs polynomial order
# plt.figure(figsize=(12,4))
# plt.plot(M_range,TrError,label='Training Errors')
# plt.plot(M_range,TeError,label="Testing Errors")
# plt.legend()
# plt.show()

########################################
#Increase the size of training data set to 100 points and repeat the experiments
########################################





########################################
#Effect of regularization
########################################
#Consider a set of lambda's. For example: lamdas = [0, 1e-7 , 1e-4, 1e-2, 1]
#Repeat the experiments, i.e., plot the prediction and error in predictions with respect to model order. Contrast these results with those obtained without regularization.
M_range=list(range(10))
lamdas = [0, 1e-7 , 1e-4, 1e-2, 1]
TrError = []
TeError = []
Train_Est=[]
Test_Est=[]
for lamda in lamdas:
    tre=[]
    tee=[]
    for M in M_range:

        #Fit Mth order polynomial i.e estimate optimal w. Use the function "PolynomialFit"
        w_est=PolynomialFit(x_train,y_train,M,lamda)

        #Predict training and testing targets
        Train_est,Test_est=PolynomialPred(w_est,x_train,x_test)
        Train_Est.append(Train_est)
        Test_Est.append(Test_est)

        #Predict errors on both training and testing data using estimated w. Use the function "PolynomialPred_Error"
        Train_error,Test_error=PolynomialPred_Error(w_est,x_train,x_test,y_train,y_test)
        tre.append(Train_error)
        tee.append(Test_error)
    TrError.append(tre)
    TeError.append(tee)

# print(len(TrError),TeError)
for i in range(len(TrError)):
    plt.figure(figsize=(12,4),dpi=300)
    plt.plot(M_range,TrError[i],label='Training Error',marker='o',color='dimgray')
    plt.plot(M_range,TeError[i],label='Testing Error',marker='o',color='cadetblue')
    plt.title(f'training error vs polynomial order and testing error vs polynomial order ,lambda={lamdas[i]}')
    plt.show()

########################################
#Increase the size of training data set to 100 points and repeat the experiments
########################################
p=2*np.pi
x=np.random.rand(100)
y=np.sin(p*x)+np.random.normal(loc=0,scale=0.1**0.5,size=100)

#Obtain train and test splits

x_train=x[:50]
y_train=y[:50]
x_test=x[50:]
y_test=y[50:]
#Iterate through range of M values
M_range=list(range(10))
x_r=np.linspace(0,1,100)
y_r=np.sin(p*x_r)
TrError = []
TeError = []
Train_Est=[]
Test_Est=[]
y_real_p=[]
for M in M_range:

    #Fit Mth order polynomial i.e estimate optimal w. Use the function "PolynomialFit"
    w_est=PolynomialFit(x_train,y_train,M,0)

    #Predict training and testing targets
    Train_est,Test_est=PolynomialPred(w_est,x_train,x_test)
    Train_Est.append(Train_est)
    Test_Est.append(Test_est)
    real,_=PolynomialPred(w_est,x_r,x_r)
    y_real_p.append(real)
    #Predict errors on both training and testing data using estimated w. Use the function "PolynomialPred_Error"
    Train_error,Test_error=PolynomialPred_Error(w_est,x_train,x_test,y_train,y_test)
    TrError.append(Train_error)
    TeError.append(Test_error)
    #Store them for plotting
# print(y_real_p)

#Plot training and testing estimates alogwith the original targets
# print(TeError,TrError)
# fig,ax=plt.subplots(nrows=5,ncols=2)

for i in range(3):

    fig,ax=plt.subplots(1,3,figsize=(12,4),dpi=300)
    fig.suptitle("$sin(2πx)$,no of points=100")
    for j in range(3):
        M=3*i+j
        ax[j].scatter(x_train,y_train,label='Training Data',color='green')
        ax[j].scatter(x_train,Train_Est[3*i+j],label='Predicted Training Data',color='lightgreen',alpha=0.5)
        ax[j].scatter(x_test,y_test,label='Testing Data',color='steelblue')
        ax[j].scatter(x_test,Test_Est[3*i+j],label="Predicted Testing Data",color='skyblue',alpha=0.5)
        ax[j].set_title(f"Plot for M={M+1}")
        ax[j].plot(x_r,y_r,color='crimson')
        ax[j].plot(x_r,y_real_p[3*i+j],color='teal')
        ax[j].set_ylim(-3,3)
        ax[j].legend()
    plt.show()

########################################
#Effect of bias regularization
########################################
#Modify the function i.e include bias
#Generate data
p=2*np.pi
x=np.random.rand(20)
y=np.sin(p*x)+np.random.normal(loc=0,scale=0.1**0.5,size=20)+5

#Obtain train and test splits

x_train=x[:10]
y_train=y[:10]
x_test=x[10:]
y_test=y[10:]
#Iterate through range of M values
M_range=list(range(10))
x_r=np.linspace(0,1,100)
y_r=np.sin(p*x_r)+5
TrError = []
TeError = []
Train_Est=[]
Test_Est=[]
y_real_p=[]
for M in M_range:

    #Fit Mth order polynomial i.e estimate optimal w. Use the function "PolynomialFit"
    w_est=PolynomialFit(x_train,y_train,M,0)

    #Predict training and testing targets
    Train_est,Test_est=PolynomialPred(w_est,x_train,x_test)
    Train_Est.append(Train_est)
    Test_Est.append(Test_est)
    real,_=PolynomialPred(w_est,x_r,x_r)
    y_real_p.append(real)
    #Predict errors on both training and testing data using estimated w. Use the function "PolynomialPred_Error"
    Train_error,Test_error=PolynomialPred_Error(w_est,x_train,x_test,y_train,y_test)
    TrError.append(Train_error)
    TeError.append(Test_error)
    #Store them for plotting
# print(y_real_p)

#Plot training and testing estimates alogwith the original targets
# print(TeError,TrError)
# fig,ax=plt.subplots(nrows=5,ncols=2)

for i in range(3):

    fig,ax=plt.subplots(1,3,figsize=(12,4),dpi=300)
    fig.suptitle("$sin(2πx)+5$,no of points=20")

    for j in range(3):
        M=3*i+j
        ax[j].scatter(x_train,y_train,label='Training Data',color='green')
        ax[j].scatter(x_train,Train_Est[3*i+j],label='Predicted Training Data',color='lightgreen',alpha=0.5)
        ax[j].scatter(x_test,y_test,label='Testing Data',color='steelblue')
        ax[j].scatter(x_test,Test_Est[3*i+j],label="Predicted Testing Data",color='skyblue',alpha=0.5)
        ax[j].set_title(f"Plot for M={M+1}")
        ax[j].plot(x_r,y_r,color='crimson')
        ax[j].plot(x_r,y_real_p[3*i+j],color='teal')
        ax[j].set_ylim(3,7)
        ax[j].legend()
    plt.show()
# for M in M_range:
#         plt.scatter(x_train,y_train,label='Training Data',color='green')
#         plt.scatter(x_train,Train_Est[M],label='Predicted Training Data',color='red',alpha=0.5)
#         plt.scatter(x_test,y_test,label='Testing Data',color='blue')
#         plt.scatter(x_test,Test_Est[M],label="Predicted Testing Data",color='yellow',alpha=0.5)
#         plt.title(f"Plot for M={M}")
#         plt.plot(x_r,y_r)
#         plt.plot(x_r,y_real_p[M])
#         plt.ylim(3,7)
#         plt.legend()
#         plt.show()
#Estimate the polynomial with and without regularization constraint
# plt.scatter(x_train,y_train,color='green')
# plt.scatter(x_test,y_test,color='yellow')

#Compare the two estimated polynomials and report the observations

```

```python
#Understanding the bias-variance trade-off
########################################
#Generate 100 data sets of noisy sinusoidal data
########################################
datasets=[]
n=100
k=25
for i in range(n):
    np.random.seed(i)
    dataset=[]
    x=np.linspace(0,2*np.pi,k)
    y=np.sin(x)+np.random.normal(0,0.1,k)
    dataset.append(x)
    dataset.append(y)
    datasets.append(dataset)
#     print(dataset)
datasets=np.array(datasets)
print(datasets.shape)
########################################
#Use regularized least squares to estimate w
########################################
def gaussian_basis(x, mu, sigma):
    return np.exp(-0.5 * ((x - mu) / sigma) ** 2)

def design_matrix(X, M, sigma):
    mu = np.linspace(0, 2*np.pi, M)
    Phi = np.column_stack([gaussian_basis(X, m, sigma) for m in mu])
    Phi = np.column_stack([np.ones_like(X), Phi])  # Add bias term
    return Phi

def regularized_least_squares(Phi, y, lam):
    xtx=np.dot(Phi.T,Phi)+lam*np.identity(M+1)
    xtx_inv = np.linalg.inv(xtx)
    w_opt = np.dot(np.dot(xtx_inv, Phi.T), y)
    return w_opt

M = 25  # Order of the linear regression model
sigma = 0.1  # Width of Gaussian basis functions

# Regularization coefficients (low, middle, high)
lambdas = [0.01, 0.1, 1.0]

estimated_curves = []

for lam in lambdas:
    w_list = []
    for X, y in datasets:
        Phi = design_matrix(X, M, sigma)
        w = regularized_least_squares(Phi, y, lam)
        w_list.append(w)
    estimated_curves.append(w_list)
estimated_curves=np.array(estimated_curves)
print((estimated_curves.shape))

########################################
#Illustrate the concept of Bias-Variance trade off
########################################
#1. Chose three different regularization coefficeints (low,middle and high)
#2. For every regularization coefficient, produce two plots: one displaying 100 estimated curves, and the other showing the mean of the estimated curves alongside the original function.
#3. For three regularization coefficients, you should have a total of six plots, meaning two plots for each regularization.
#4. Using the six plots above, describe the bias-variance trade-off.

# Plot 100 estimated curves and mean of the estimated curves alongside the original function
fig,ax=plt.subplots(3,2,figsize=(15,12))
for i in range(len(lambdas)):
    ax[i][0].set_title(f"regularization coefficient:{lambdas[i]}")
    for j in range(n):
        ax[i][0].plot(X,np.dot(design_matrix(X,M,sigma),estimated_curves[i][j]),color='lightslategrey',alpha=0.3)
    ax[i][0].set_xlabel("X")
    ax[i][0].set_ylabel("Y")

    ax[i][1].set_title(f"Mean of estimated curves vs. Original function (Regularization coefficient: {lambdas[i]})")
    mean_curve = np.mean([np.dot(design_matrix(X, M, sigma), w) for w in estimated_curves[i]], axis=0)
    ax[i][1].plot(X, mean_curve, label='Mean estimated curve', color='dimgrey')
    original_function = np.sin(X)
    ax[i][1].plot(X, original_function, label='Original function', linestyle='--',color='gold')
    ax[i][1].legend()
    ax[i][1].set_xlabel('X')
    ax[i][1].set_ylabel('y')
plt.tight_layout()
plt.show()
```

```python
#Real world problems
#Part(2): Breast Cancer Wisconsin (Prognostic) Data Set

from google.colab import drive
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt

drive.mount('/content/drive',force_remount=True)
#Download the data from the given link and keep it in the following location
datapath = "/content/drive/MyDrive/PRML_Assignments/Assignment_2/BreastCancer/wpbc.data"
data = []
with open(datapath) as fp:
    Lines = fp.readlines()
    for line in Lines:
      data.append(line.strip().split(',')[2:33])
data = np.array(data)
data = data.astype('float32')
#Shuffle the data
np.random.shuffle(data)
n=int(len(data)*0.5)
#Train and test splits
train=data[:n]
test=data[n:]
X_train = []
Y_train=[]
X_test=[]
Y_test=[]
for i in range(len(train)):
  arr=train[i][1:]
  arr=np.insert(arr, 0, 1)
  X_train.append(arr)
  Y_train.append(train[i][0])
for i in range(len(test)):
  arr=test[i][1:]
  arr=np.insert(arr, 0, 1)
  X_test.append(arr)
  Y_test.append(train[i][0])
X_train=np.array(X_train)
Y_train=np.array(Y_train)
X_test=np.array(X_test)
Y_test=np.array(Y_test)

c=len(X_train[0])
lamda=0.1
xtx=np.dot(X_train.T,X_train)+lamda*np.identity(c)
xtx_inv = np.linalg.inv(xtx)
w_opt = np.dot(np.dot(xtx_inv, X_train.T), Y_train)
print(w_opt.shape)
Y_test_pred=np.dot(X_test,w_opt)
x=np.arange(1,len(Y_test)+1)
plt.figure(figsize=(12,5),dpi=300)
plt.scatter(x,Y_test,label="testing data",color='violet',marker='o')
plt.scatter(x,Y_test_pred,label='predicted testing data',color='dimgrey',marker='o',alpha=0.5)
plt.xlabel('index of data')
plt.ylabel('value')
plt.legend()
plt.show()
```
